```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def act_pow(x, p=0.5):
    return np.power(x, p)

def act_tanh(x, mid=0.5, sat=4):
    return 1 / (1 + np.exp(-2 * sat * (x - mid)))

def act_tanh2(x, mid=0.5, sat=4):
    up_range = 1.0 - mid
    lo_range = mid
    up_idx = x > mid
    lo_idx = x < mid
    y = x.copy()
    y[up_idx] = 1 / (1 + np.exp(- (sat / up_range) * (x[up_idx] - mid)))
    y[lo_idx] = 1 / (1 + np.exp(- (sat / lo_range) * (x[lo_idx] - mid)))
    return y

def act_tanh3(x, mid=0.5, sat=4):
    y = act_tanh(x, mid, sat)
    return (y - np.min(y)) / (np.max(y) - np.min(y))

def act_tanh4(x, mid=0.5, sat=4):
    y = act_tanh2(x, mid, sat)
    return (y - np.min(y)) / (np.max(y) - np.min(y))
```

```python
plt.plot(x, x, label=f'y=x')
for p in np.linspace(0.1, 0.9, 5):
    y = act_pow(x, p)
    plt.plot(x, y, label=f'pow-{p:.2f}')
plt.legend()
plt.show()
```

```python
plt.plot(x, x, label=f'y=x')
for m in np.linspace(0.1, 0.9, 5):
    y = act_tanh(x, mid=m)
    plt.plot(x, y, label=f'tanh-{m:.2f}-4.0')
plt.legend()
plt.show()
```

```python
plt.plot(x, x, label=f'y=x')
for s in np.linspace(1.0, 5.0, 5):
    y = act_tanh(x, sat=s)
    plt.plot(x, y, label=f'tanh-0.5-{s:.2f}')
plt.legend()
plt.show()
```

```python
plt.plot(x, x, label=f'y=x')
for m in np.linspace(0.1, 0.9, 5):
    y = act_tanh2(x, mid=m)
    plt.plot(x, y, label=f'tanh2-{m:.2f}-4.0')
plt.legend()
plt.show()
```

```python
plt.plot(x, x, label=f'y=x')
for m in np.linspace(0.1, 0.9, 5):
    y = act_tanh3(x, mid=m)
    plt.plot(x, y, label=f'tanh-{m:.2f}-4.0')
plt.legend()
plt.show()
```

```python
plt.plot(x, x, label=f'y=x')
for s in np.linspace(1.0, 5.0, 5):
    y = act_tanh3(x, sat=s, mid=0.3)
    plt.plot(x, y, label=f'tanh-0.3-{s:.2f}')
plt.legend()
plt.show()
```

```python
plt.plot(x, x, label=f'y=x')
for m in np.linspace(0.1, 0.9, 5):
    y = act_tanh4(x, mid=m)
    plt.plot(x, y, label=f'tanh2-{m:.2f}-4.0')
plt.legend()
plt.show()
```

```python
plt.plot(x, x, label=f'y=x')
for s in np.linspace(1.0, 5.0, 5):
    y = act_tanh4(x, sat=s, mid=0.3)
    plt.plot(x, y, label=f'tanh2-0.3-{s:.2f}')
plt.legend()
plt.show()
```
